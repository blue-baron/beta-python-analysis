#### Remove NaN values from open text field columns to avoid type errors.

```python
# Variable for storing open text field column names.
# Note that col names must be unique or errors will occur down the track
open_text_cols = ['Other, please specify', '* What were your reasons for choosing that lodgment method?', 'Is there anything you would improve on this design?', 'Other', 'Give us your feedback on how we can improve the tax time lodgment pages.','Other, please specify (2)']
open_text_cols
```

```python
# Define function to remove NaN values from all necessary columns
def remove_nan(dataframe, columns):
    for item in columns:
        dataframe[item] = dataframe[item].fillna('nan')
        
remove_nan(raw_df, open_text_cols)
len(raw_df)
```

#### Define function to check for testers identified in the open text questions.

```python
# Function to check for a partial string in the selected columns.
# LDC 18/5/2017: Fixed to include all testers not just those found from the last col of open_text_columns
def remove_testers(dataframe, fragment, columns):
    for i, col in enumerate(columns):
        temp = dataframe[dataframe[col].str.contains('(?i)'+fragment) == True]
        if i == 0:
            test_df = temp
        else:
            # Union merge the dataframes
            test_df = pd.merge(test_df, temp, how='outer')
       
    return test_df

testers_df = remove_testers(raw_df, 'test', open_text_cols)
```

```python
# Print out the testers to check the comments are genuine tester comments.
# Extend width so that to view all columns

pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
testers_df[['Participant No.']+open_text_cols]
```

```python
# Remove participants that have falsely been id'd as testers
# Change these to fit the situation
actual_testers = testers_df.loc[(testers_df['Participant No.'] != 'Participant 93') 
               & (testers_df['Participant No.'] != 'Participant 249')
               & (testers_df['Participant No.'] != 'Participant 248')
               & (testers_df['Participant No.'] != 'Participant 395')
               & (testers_df['Participant No.'] != 'Participant 413')]

actual_testers
```

```python
# Remove the actual testers from the raw data set to produce the cleaned set
# Remove by XOR deletion aka combine sets and delete overlap
pd.set_option('display.max_rows', 500)

doubled_up = raw_df.append(actual_testers)
tester_free_df = doubled_up.drop_duplicates(subset=None, keep=False, inplace=False)
len(tester_free_df) # sanity check
```

#### Looking at Total Time spent for outliers, rather than avg time.

```python
# Convert string data to integer.
times_temp = []
for i in tester_free_df['Total Time Spent']:
    times_temp.append(int(i[:-5]))

# Ignore the warning that will be generated by reassignment (see for justification: http://stackoverflow.com/questions/20625582/how-to-deal-with-settingwithcopywarning-in-pandas)
pd.options.mode.chained_assignment = None 
tester_free_df['Total Time Spent'] = times_temp

tester_free_df['Total Time Spent'].describe()
```

```python
# Calculate 2 std below the mean
times_mean = tester_free_df['Total Time Spent'].mean()
times_std = tester_free_df['Total Time Spent'].std()
    
mean_less_2xStd = times_mean - (times_std * 2)

tester_free_df[tester_free_df['Total Time Spent'] < mean_less_2xStd]
```

```python
# Calculate 2 std above the mean
mean_plus_2xStd = times_mean + (times_std * 2)
mean_plus_2xStd
# how many instances are above 2 std? (sanity check)
tester_free_df[tester_free_df['Total Time Spent'] > mean_plus_2xStd]
```

```python
# drop values from cleaned data that are 2 std above the mean
clean_df = tester_free_df[tester_free_df['Total Time Spent'] <= mean_plus_2xStd]
clean_df
```

#### Export clean data to csv.

```python
# Export spreadsheet with only full task complete information.
clean_df.to_csv('ttlp_test_clean.csv',',') # export to csv
```
