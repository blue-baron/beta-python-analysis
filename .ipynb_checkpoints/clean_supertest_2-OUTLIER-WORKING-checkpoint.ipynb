#### Import packages.

```python
import pandas as pd
```

#### Import raw data csv from Loop11.

```python
# Create variable and create DataFrame from the raw data exported from testing tool.
raw_df = pd.read_csv('supertest2_RAW.csv', encoding='latin-1')
```

```python
# List the DataFrame Column names.
list(raw_df.columns.values)
```

```python
# Checking the number of participants in the raw data.
len(raw_df)
```

#### Remove NaN values from columns to avoid type errors.

```python
# Variable for storing open text field column names.
open_text_cols = ['Other, please specify', 'Other, please specify.1', 'Other, please specify.2', 'Other, please specify.3']
```

```python
# Define function to remove NaN values from all necessary columns
def remove_nan(dataframe, columns):
    for item in columns:
        raw_df[item] = raw_df[item].fillna('nan')
        
remove_nan(raw_df, open_text_cols)
```

#### Define function to check for testers identified in the open text questions.

```python
# Function to check for a partial string in the selected columns.
def remove_testers(fragment, columns):
    for item in columns:
        test_df = raw_df[raw_df[item].str.contains('(?i)'+fragment) == True]    
    return test_df

testers_df = remove_testers('test', open_text_cols)
```

```python
# Print out the testers to check the comments are genuine tester comments.
testers_df
```

```python
# Remove participants identified as testers.
clean_df = raw_df[raw_df['Participant No.'] != 'Participant 1']
clean_df = clean_df[clean_df['Participant No.'] != 'Participant 6']

clean_df
```

```python
len(clean_df)
```

#### Remove incomplete task data (Missing task completion rates caused by Loop11 malfunction).

```python
# Remove impartial completes (check for null data in task completion rates fields).
fullcompletes_df = clean_df[clean_df['2. Task. Task 1 - Employer details'].notnull() & clean_df['7. Task. Task 2 - Employee details'].notnull() & clean_df['17. Task. Task 3 Understanding of results'].notnull() ] 

# Check number of participants in final sample.
len(fullcompletes_df)
```

#### Export clean data to csv.

```python
fullcompletes_df.to_csv('supertest_fullcompletes_2.csv',',') # export to csv
```

```python
clean_df.to_csv('supertest_partialcompletes_2.csv',',') # export to csv
```
