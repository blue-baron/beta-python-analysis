```python
from IPython.display import display, HTML

import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def autolabel(rects):
    # attach some text labels
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.05*height,
                '%d' % int(height),
                ha='center', va='bottom')
```

```python
# DATA FOR PLOTTING
N = 5
menMeans = (20, 35, 30, 35, 27)
menStd = (2, 3, 4, 1, 2)
womenMeans = (25, 32, 34, 20, 25)
womenStd = (3, 5, 2, 3, 3)
catMeans = (30, 20, 40, 10, 15)
catStd = (3, 2, 2, 4, 5)

# X AXIS FORMATTING (bar placement and width)
ind = np.arange(N)  # the x locations for the groups
width = 0.25

# CREATE INSTANCE OF GRAPH
fig, ax = plt.subplots()

men = ax.bar(ind, menMeans, width, color='b', yerr=menStd, ecolor="k")
women = ax.bar(ind + width, womenMeans, width, color='g', yerr=womenStd, ecolor="k")
cats = ax.bar(ind + width + width, catMeans, width, color='y', yerr=catStd, ecolor="k")

# add some text for labels, title and axes ticks
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(ind + width + width)
ax.set_xticklabels(('G1', 'G2', 'G3', 'G4', 'G5'))

ax.legend((men[0], women[0], cats[0]), ('Men', 'Women', 'Cats'))

autolabel(men)
autolabel(women)
autolabel(cats)
```
