#### Outliers

```python
# Convert 'Total Time Spent' string type to integer.
times_temp = []
for i in fullcompletes_df['Total Time Spent']:
    times_temp.append(int(i[:-5]))

    
# Create DataFrame and use describe() function to get stats.    
times_df = pd.DataFrame(times_temp)
times_df.columns = ['Total Time Spent']
times_df['Total Time (minutes)'] = times_df['Total Time Spent'] / 60
times_df.describe()
    
```

```python
times_df.sort_values('Total Time Spent', ascending=True)
```

```python
# Reset index of fullcompletes so it matches times_df (and pandas.update() will for correctly).
fullcompletes_df.reset_index(inplace=True, drop=True)
```

```python
times_mean = times_df['Total Time Spent'].mean()
times_std = times_df['Total Time Spent'].std()

# Calculate two std below the mean
mean_less_2xStd = times_mean - (times_std * 2)

# Calculate two std above the mean
mean_plus_2xStd = times_mean + (times_std * 2)

# Calculate three std above the mean
mean_plus_3xStd = times_mean + (times_std * 3)

print('Two Std below the mean: ', mean_less_2xStd)
print('Two Std above the mean: ', mean_plus_2xStd)
print('Three Std above the mean: ', mean_plus_3xStd)
```

```python
# Update the Cleaned DataFrame with the number values.
fullcompletes_df.update(times_df[['Total Time Spent']], raise_conflict=False)
fullcompletes_df
```

```python
# Function to remove outliers based on very fast or very slow times.
def remove_outliers(df, col, min, max):
    outliers_df = df[df[col] >= min]
    outliers_df = df[df[col] <= max]
    return outliers_df
```

```python
remove_outliers_df = remove_outliers(fullcompletes_df, 'Total Time Spent', mean_less_2xStd, mean_plus_2xStd)
len(remove_outliers_df)
```
