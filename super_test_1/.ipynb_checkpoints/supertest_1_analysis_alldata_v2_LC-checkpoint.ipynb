```python
import pandas as pd
import numpy as np
import math
```

```python
### THE FOLLOWING CODE IS FOR GENERATING THE DATA FOR BOTH TASKS (JOINT SAMPLE SIZE, N = 51)
### IF YOU WOULD LIKE THE TASK SPECIFIC DATA SEE BELOW.
clean_df = pd.read_csv('clean_supertest_1_all_data.csv') # Read data in
```

```python
# Define sample size for future use in the script as a variable called N
N = len(clean_df)
```

```python
list(clean_df.columns.values) # For reference later when picking out data in specific column
```

```python
# Create results dataframe for all results :
results = []

tasks = ['2. Task. Estimate your super', '8. Task. Report unpaid super']
for t in tasks:
    results.append(pd.DataFrame(clean_df[t].value_counts()))
    results.append(pd.DataFrame(clean_df[t].value_counts(normalize=True)))

results = pd.concat(results, axis=1)
results.index.names = ['Test Result']
results.columns = ['Task 2 (N)', 'Task 2 (%)','Task 8 (N)', 'Task 8 (%)']
#results.to_csv('Documents/supertest_1_task_results.csv',',') # export to csv

results = results.sort_index(ascending=False) # or results.reindex(['success', 'fail', 'abandon'])
results
```

```python
# For small sample size adjustment
# Adjust the sample size, N
def adjust_n(N):
    return N+(1.96**2)

# Adjust the proportion of successes, p
def adjust_p(p, N):
    return (p+1.92)/(N+3.84)

T2_p = results['Task 2 (N)']['success']
T8_p = results['Task 8 (N)']['success']

# To calculate the confidence intervals the formula used with these values is:
#    CI_95 = 1.96 * SQRT((adjusted_p * (1 - adjusted_p))/adjusted_n)
adjust_n(N), adjust_p(T2_p,N), adjust_p(T8_p,N)

T2_CI = 1.96 * np.sqrt((adjust_p(T2_p,N) * (1 - adjust_p(T2_p,N)))/adjust_n(N))
T8_CI = 1.96 * np.sqrt((adjust_p(T2_p,N) * (1 - adjust_p(T8_p,N)))/adjust_n(N))

# for reference:
T2_CI, T8_CI
```

```python
employee_breakdown = pd.DataFrame(clean_df['Are you an ATO employee?'].value_counts())
employee_breakdown
```

```python
# Create resutls for ATO versus non-ATO
ATO_results = []
for t in tasks:
    ATO_results.append(pd.DataFrame(clean_df[t].groupby(clean_df['Are you an ATO employee?']).value_counts()))
    ATO_results.append(pd.DataFrame(clean_df[t].groupby(clean_df['Are you an ATO employee?']).value_counts(normalize=True)))
    
ATO_results = pd.concat(ATO_results, axis=1)
ATO_results.index.names = ['ATO Employee','Test Result']
ATO_results.columns = ['Task 2 (N)', 'Task 2 (%)','Task 8 (N)', 'Task 8 (%)']
ATO_results = ATO_results.sort_index(ascending=False) # or results.reindex(['success', 'fail', 'abandon'])
#ATO_results.to_csv('Documents/supertest_1_task_results_employee_breakdown.csv',',') # export to csv
ATO_results
```

```python
# for calculating the CIs
ATO_N = employee_breakdown['Are you an ATO employee?']['Yes']
nonATO_N = employee_breakdown['Are you an ATO employee?']['No']
task_cols = ['Task 2 (N)', 'Task 8 (N)']

# create empty lists to collect the confidence intervals in 
nonATO_cis = [] 
ATO_cis = [] 

for t in task_cols:
    temp = adjust_p(ATO_results[t]['No']['success'],nonATO_N)
    nonATO_cis.append(1.96 * np.sqrt((adjust_p(temp,nonATO_N) * (1 - adjust_p(temp,nonATO_N)))/adjust_n(nonATO_N)))

for t in task_cols:
    temp = adjust_p(ATO_results[t]['Yes']['success'],ATO_N)
    ATO_cis.append(1.96 * np.sqrt((adjust_p(temp,ATO_N) * (1 - adjust_p(temp,ATO_N)))/adjust_n(ATO_N)))

# For reference
nonATO_cis, ATO_cis
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
%matplotlib inline

### For plotting graphs
def grapher_errorbar(df, cols, title, cis, labels):
    plot_df = df[cols]
    colours=['#9d9d9c','#074f5f','#42bac7']
    ax = plot_df.plot(kind='bar', legend=False, color=colours, edgecolor = "none", rot = 0, yerr=cis)
    ax.set_xticklabels(list(plot_df[cols].index))
    ax.set_ylim([0,1])
    ax.yaxis.set_major_formatter(FuncFormatter(lambda y, _: '{:.0%}'.format(y)))
    ax.set_title(title)
    if not(not labels):
        ax.legend(labels,loc='best')
    plt.savefig("super-1-graphs/"+title+".svg")
    
grapher_errorbar(results, ['Task 2 (%)'], 'Task 2 completion rates (%)', [T2_CI,T2_CI,T2_CI],[])
grapher_errorbar(results, ['Task 8 (%)'], 'Task 8 completion rates (%)', [T8_CI,T8_CI,T8_CI],[])

errors = pd.DataFrame(np.array([[T2_CI,T8_CI],[T2_CI,T8_CI],[T2_CI,T8_CI]]), index=['success', 'fail', 'abandon'], columns=['Task 2 (%)','Task 8 (%)'])
grapher_errorbar(results, ['Task 2 (%)','Task 8 (%)'], 'Task completion rates (%)',errors,['Task 2','Task 8'])
```

```python
### For plotting graphs with stacked data sets
def grapher_stacked(df, cols, title, labels, cis):
    plot_df = df[cols].unstack(level=0).sort_index(ascending=False)
    colours=['#42bac7','#074f5f','#9d9d9c']
    ax = plot_df.plot(kind='bar', legend=True, color=colours, edgecolor = "none", rot = 0, yerr=cis)
    ax.set_xticklabels(list(plot_df.index))
    ax.set_ylim([0,1])
    ax.yaxis.set_major_formatter(FuncFormatter(lambda y, _: '{:.0%}'.format(y)))
    ax.set_title(title)
    ax.legend(labels,loc='best')
    plt.savefig("super-1-graphs/"+title+".svg")


# Convert the CIs into a dataframe for plotting for both Tasks 2 and 8
T2_errors = pd.DataFrame(np.array([[nonATO_cis[0],ATO_cis[0]],[nonATO_cis[0],ATO_cis[0]],[nonATO_cis[0],ATO_cis[0]]]), 
                          index=['success', 'fail', 'abandon'], columns=['No','Yes'])
T2_errors.index.names = ['test']
T2_errors

T8_errors = pd.DataFrame(np.array([[nonATO_cis[1],ATO_cis[1]],[nonATO_cis[1],ATO_cis[1]],[nonATO_cis[1],ATO_cis[1]]]), 
                          index=['success', 'fail', 'abandon'], columns=['No','Yes'])
T8_errors.index.names = ['test']
T8_errors

# Plot them:
grapher_stacked(ATO_results, 'Task 2 (%)', 'Task 2 completion rates ATO vs Non-ATO (%)', ['non-ATO employee','ATO employee'], T2_errors)
grapher_stacked(ATO_results, 'Task 8 (%)', 'Task 8 completion rates ATO vs Non-ATO (%)', ['non-ATO employee','ATO employee'], T8_errors)
```
